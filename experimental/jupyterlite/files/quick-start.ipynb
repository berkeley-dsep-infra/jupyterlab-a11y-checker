## Getting Started with the Accessibility Checker

This notebook contains intentional accessibility issues for you to try out the **jupyterlab-a11y-checker** extension.

### How to use:
1. Click the **accessibility icon** (blue universal access symbol) in the right sidebar
2. Click **"Analyze Notebook"**
3. Review the detected issues and try the suggested fixes

---

## Issue 1: Missing Top-Level Heading

This notebook starts with H2 instead of H1. The accessibility checker should flag this as a heading hierarchy issue.

#### Issue 2: Skipped Heading Level

This cell jumps from H2 to H4, skipping H3. This makes navigation harder for screen reader users.
