### Logo

Inspired by famed type designer Frederic Goudy’s Californian typeface — which was created more than 85 years ago for the University of California Press — and its sibling font, Deepdene, our logo is designed to be clear, confident and curious. With its bold letterforms and upward-facing “e’s,” we are reminded of UC Berkeley’s mission as a public research university. The terminal at the end of the “y,” is a nod to the terminal in the “c” of the Cal script, and is designed to create a subtle, but important visual connection between the academic and athletic identities.
